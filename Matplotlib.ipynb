```python
#Scatter Graph
import matplotlib.pyplot as plt

# Data
x = [1, 2, 3, 4]
y = [2.2, 3.3, 4.5, 6.7]
colors = ['r', 'y', 'b', 'g']  # Color for each point
sizes = [100, 2000, 10, 20]       # Size of each point

# Create scatter plot
plt.scatter(x, y, c=colors, s=sizes, alpha=0.7, edgecolor='k')  # Use colors and sizes


t=plt.colorbar()
t.set_label("Colorbar",fontsize=15)
# Title and labels
plt.title('Scatter Plot', fontsize=20)
plt.xlabel("Day", fontsize=15)
plt.ylabel("No.", fontsize=15)

# Show the plot
plt.grid(True)  # Optional: Add grid for better readability
plt.show()
```

```python
#stack plot
import numpy as np
import matplotlib.pyplot as plt

# Data for three categories over time
days = [1, 2, 3, 4, 5]
category1 = [3, 4, 2, 5, 6]
category2 = [1, 2, 3, 4, 5]
category3 = [2, 1, 3, 2, 1]

# Plotting Stack Plot
plt.stackplot(days, category1, category2, category3, labels=['Category 1', 'Category 2', 'Category 3'], colors=['blue', 'orange', 'green'])

plt.xlabel('Days')
plt.ylabel('Values')
plt.title('Stack Plot Example')
plt.legend(loc='upper left')
plt.show()
```

```python
#Fill_Between Plot
x=np.linspace(0,10,100)
y1=np.sin(x)
y2=np.sin(x)+0.5

plt.plot(x,y1,label="slim_curves",color='blue')
plt.plot(x,y2,label="thick_curves",color='red')
plt.fill_between(x, y1, y2, color='gray', alpha=0.5)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Fill Between Plot')
plt.legend(loc=2)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

# Creating subplots (2 rows, 1 column)
plt.figure(figsize=(8, 6))

# First subplot
plt.subplot(2, 1, 1)
plt.plot(x, y1, color='blue', label='sin(x)')
plt.legend()
plt.title('Sine Wave')

# Second subplot
plt.subplot(2, 1, 2)
plt.plot(x, y2, color='red', label='cos(x)')
plt.legend()
plt.title('Cosine Wave')

plt.tight_layout() # Prevents overlap
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 10, 10)
y = np.sin(x)

plt.plot(x, y, color='blue')

# Custom X and Y tick labels
plt.xticks(np.arange(0, 11, 2), ['Zero', 'Two', 'Four', 'Six', 'Eight', 'Ten'])
plt.yticks(np.arange(-1, 1.1, 0.5))

plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('Customized Ticks Example')
plt.grid(True)
plt.show()
```
