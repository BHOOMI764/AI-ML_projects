```python
y_1=sns.load_dataset("penguins")
y_1
```

```python
sns.lineplot(x="bill_length_mm", 
             y="flipper_length_mm", 
             data=y_1, 
             hue="sex", 
             style="sex", 
             markers=["o", "*"])
plt.grid(True)
plt.title("Python Data Visualization")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load example dataset
df = sns.load_dataset('penguins')

# Plot using displot
sns.displot(df['bill_depth_mm'])
plt.show()


sns.displot(df['flipper_length_mm'],bins=[170,180,190,200,210,220,230,240],color='g')
plt.show()

sns.barplot(x=df.island , y = df.bill_length_mm, hue="sex",data =df)
plt.show()

sns.scatterplot(x='bill_length_mm', y='bill_depth_mm', data=df, hue="sex")
plt.title('Bill Length vs Bill Depth')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
df = sns.load_dataset('penguins')

# Basic count plot
sns.countplot(x='species',hue="sex", data=df)

plt.title('Count of Each Penguin Species')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the correct dataset
df = sns.load_dataset('tips')

# Create a violin plot
sns.violinplot(x='day', y='total_bill', data=df)

plt.title('Total Bill Distribution by Day')
plt.show()
```

```python
sns.pairplot(df,hue='sex')
plt.show()
```
